```python
#Dependencies
import pandas as pd
```

```python
#Reading in files
df_2013 = pd.read_csv(
    "2013-07 - Citi Bike trip data.csv")
df_2018 = pd.read_csv(
    "201807-citibike-tripdata.csv")
df_2023 = pd.read_csv(
    "202307-citibike-tripdata.csv")
```

```python
#Viewing data
df_2013.head()
```

```python
#Collecting data on the number of times someone started at the given station
start_value_counts_df_2013 = df_2013['start station name'].value_counts().reset_index()
start_value_counts_df_2013 = start_value_counts_df_2013.rename(columns={'start station name': 'start_station_name', 'count': 'starts_2013'})

start_value_counts_df_2018 = df_2018['start station name'].value_counts().reset_index()
start_value_counts_df_2018 = start_value_counts_df_2018.rename(columns={'start station name': 'start_station_name', 'count': 'starts_2018'})

start_value_counts_df_2023 = df_2023['start_station_name'].value_counts().reset_index()
start_value_counts_df_2023 = start_value_counts_df_2023.rename(columns={'count': 'starts_2023'})
```

```python
#Collecting coordinates for each starting station
group_df = df_2023[['start_station_name', 'start_lat','start_lng']]
group_df = group_df.groupby('start_station_name', as_index=False).mean()
group_df
```

```python
#Adding the number of starts per year to the dataframe
result_df = pd.merge(group_df, start_value_counts_df_2013, on='start_station_name', how='left')
result_df = pd.merge(result_df, start_value_counts_df_2018, on='start_station_name', how='left')
result_df = pd.merge(result_df, start_value_counts_df_2023, on='start_station_name', how='left')

result_df
```

```python
#Collecting data on the number of times someone ended at the given station
end_value_counts_df_2013 = df_2013['end station name'].value_counts().reset_index()
end_value_counts_df_2013 = end_value_counts_df_2013.rename(columns={'end station name': 'end_station_name', 'count': 'ends_2013'})

end_value_counts_df_2018 = df_2018['end station name'].value_counts().reset_index()
end_value_counts_df_2018 = end_value_counts_df_2018.rename(columns={'end station name': 'end_station_name', 'count': 'ends_2018'})

end_value_counts_df_2023 = df_2023['end_station_name'].value_counts().reset_index()
end_value_counts_df_2023 = end_value_counts_df_2023.rename(columns={'count': 'ends_2023'})
```

```python
#Collecting coordinates for each ending station
group_df = df_2023[['end_station_name', 'end_lat','end_lng']]
group_df = group_df.groupby('end_station_name', as_index=False).mean()
group_df['start_station_name'] = group_df['end_station_name']

group_df
```

```python
#Combining the starting and ending dataframes and adding the number of ends per year to the dataframe
result_df = pd.merge(result_df, group_df, on='start_station_name', how='right')
result_df = pd.merge(result_df, end_value_counts_df_2013, on='end_station_name', how='left')
result_df = pd.merge(result_df, end_value_counts_df_2018, on='end_station_name', how='left')
result_df = pd.merge(result_df, end_value_counts_df_2023, on='end_station_name', how='left')

result_df
```

```python
#Exporting to csv
result_df.to_csv('cleanedBike.csv', index=False)
```
